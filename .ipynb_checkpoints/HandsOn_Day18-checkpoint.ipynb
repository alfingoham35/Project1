```python
# Data Prep
housing_raw = pd.read_csv("housing_data.csv", parse_dates=["period_begin", "period_end"])

ca_housing = (
    housing_raw
    .loc[housing_raw["region_name"].str.contains("CA")]
    .assign(region_name = housing_raw["region_name"].str[0:-11])
    .query("region_name in ['San Francisco', 'Los Angeles', 'San Diego']")
)

ca_housing_markets = (
    ca_housing
    .set_index("period_begin")
    .loc[:, ["region_name", "median_active_list_price"]]
    .sort_index()
    .pivot_table(
        index="period_begin",
        columns="region_name",
        values="median_active_list_price"
    )
   .resample("Q")
   .mean()
)

inventory = (
    ca_housing
    .loc[:, ["period_begin", "region_name", "inventory"]]
    .pivot_table(
        index="period_begin",
        columns="region_name",
        values="inventory"
    )
)

inventory.head(3)

```

```python
fig, ax = plt.subplots()

ax.stackplot(
    inventory.index, 
    inventory["San Diego"], 
    inventory["Los Angeles"], 
    inventory["San Francisco"]
)

ax.set_title("Housing Inventory Select Markets 2017-2022")
ax.set_ylabel("Inventory")
ax.legend(["San Diego", "Los Angeles", "San Francisco"])
```

```python
sf_dual = (ca_housing
          .loc[ca_housing["region_name"] == "San Francisco", 
              ["period_begin", "inventory", "median_active_list_price"]]
          .set_index("period_begin")
          .sort_index()
    )

sf_dual.head(1)
```

```python
fig, ax = plt.subplots()

ax.plot(
    sf_dual.index, 
    sf_dual["median_active_list_price"], 
    label="Price",
    c="orange"
)

ax.plot(
    sf_dual.index, 
    sf_dual["inventory"], 
    label="Inventory",
    c="green"
)

ax.set_title("Prices Drop as Inventory Rises")
ax.set_ylabel("Price (Millions)")
ax.set_ylim(0, 1600000)
ax.legend()
```

```python
fig, ax = plt.subplots()

fig.suptitle("Surge in Inventory Depresses Prices")

ax.plot(sf_dual.index, 
        sf_dual["median_active_list_price"], 
        label="Price",
        c="orange")

ax.set_ylabel("Price (Millions)")
ax.set_ylim(0, 1600000)

ax2 = ax.twinx()

ax2.plot(sf_dual.index, 
         sf_dual["inventory"], 
         label="Inventory",
         c="green")

ax2.set_ylabel("Inventory")
ax2.set_ylim(0, 2500)

fig.legend(bbox_to_anchor=(.9, .27))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame 'ca_housing' is already loaded

# Normalize the data to make the bars sum to 100%
total = ca_housing['San Francisco'] + ca_housing['Los Angeles']
ca_housing['San Francisco'] = ca_housing['San Francisco'] / total
ca_housing['Los Angeles'] = ca_housing['Los Angeles'] / total

# Plotting the 100% stacked bar chart
plt.figure(figsize=(10, 6))

plt.bar(ca_housing.index, ca_housing['San Francisco'], label='San Francisco', color='Grey')
plt.bar(ca_housing.index, ca_housing['Los Angeles'], label='Los Angeles', color='Blue', bottom=ca_housing['San Francisco'])

# Set titles and labels
plt.title('100% Stacked Bar Chart: Los Angeles and San Francisco Median Home Prices')
plt.xlabel('Date')
plt.ylabel('Percentage')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for sales totals (as an example)
sales_totals = {
    'San Francisco': 50000,
    'San Diego': 30000,
    'Los Angeles': 150000
}
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Create a pie chart
ax.pie(sales_totals.values(),
       labels=sales_totals.keys(),
       startangle=90,
       autopct='%1.1f%%',
       pctdistance=0.85)  # Adjust distance of percentage labels

# Create the "hole" by adding a white circle in the center
donut_hole = plt.Circle((0, 0), 0.70, fc='white')  # 0.70 defines the size of the hole
fig.gca().add_artist(donut_hole)

# Add a title to the chart
ax.set_title('Share of Home Sales Select CA Markets')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame with diamond data
data = {
    'carat': [0.23, 0.21, 0.23, 0.29, 0.31, 0.45, 0.50, 0.75, 1.00, 1.25],
    'cut': ['Ideal', 'Premium', 'Good', 'Premium', 'Good', 'Fair', 'Ideal', 'Good', 'Fair', 'Ideal'],
    'color': ['E', 'E', 'E', 'VS2', 'I', 'G', 'D', 'I', 'H', 'G'],
    'clarity': ['SI2', 'SI1', 'VS1', 'VS2', 'SI2', 'VS1', 'SI1', 'VS2', 'SI1', 'VS2'],
    'depth': [61.5, 59.8, 56.9, 62.4, 63.3, 61.2, 62.0, 60.1, 59.7, 61.3],
    'table': [55.0, 61.0, 65.0, 58.0, 58.0, 57.0, 58.0, 59.0, 58.0, 56.0],
    'price': [326, 326, 327, 334, 335, 525, 550, 675, 850, 1000]
}

# Create the DataFrame
df = pd.DataFrame(data)
df
```

```python

# Create the bubble chart
plt.figure(figsize=(10, 6))

# Plotting carat on x, price on y, and depth for bubble size
plt.scatter(df['carat'], df['price'], s=df['depth']*10, color='c', alpha=0.6, edgecolors="w", marker='o')

# Set the title and labels
plt.title('Diamond Price vs. Carat Weight (Bubble Chart)', fontsize=14)
plt.xlabel('Carat Weight', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)

# Show the plot
plt.show()
```
